```python
percentile25 = df['Age'].quantile(0.25)
percentile75 = df['Age'].quantile(0.75)
```

```python
IQR = percentile75 - percentile25
IQR
```

```python
upper_limit = percentile75 + 1.5*IQR
upper_limit
```

```python
lower_limit = percentile25 - 1.5*IQR
lower_limit
```

```python
ol[ol['Age']>upper_limit]
```

```python
df[ol['Age']<lower_limit]
```

```python
newol = ol[ol['Age']<upper_limit]
```

```python
# Comparision 
plt.figure(figsize=(15,7))

plt.subplot(222)
sns.boxplot(x=ol['Age'])

plt.subplot(224)
sns.boxplot(x=newol['Age'])
plt.show
```

```python
new_df_cap = ol.copy() 
```

```python
new_df_cap['Age']=np.where(
    new_df_cap['Age']>upper_limit,upper_limit,

    np.where(
        new_df_cap['Age']<lower_limit,lower_limit,
        new_df_cap['Age']
    )
)
```

```python
# Comparision 
plt.figure(figsize=(15,7))

plt.subplot(222)
sns.boxplot(x=ol['Age'])

plt.subplot(224)
sns.boxplot(x=new_df_cap['Age'])
plt.show
```
