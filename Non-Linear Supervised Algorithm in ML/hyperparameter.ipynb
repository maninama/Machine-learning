```python
import pandas as pd 
```

```python
dataset = pd.read_csv(r"D:\py_machine_learning\datasets\polynomial.csv")
dataset.head()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.42,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(criterion='squared_error',splitter='best',max_depth = 8)
dt.fit(x_train,y_train)
dt.score(x_train,y_train)*100, dt.score(x_test,y_test)*100  
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
df = param_grid = {
    "criterion": ["squared_error", "friedman_mse", "absolute_error", "poisson"],
    "splitter": ["best", "random"],
    "max_depth": [i for i in range(2, 20)]
}
```

```python
rd = RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=df,n_iter=20)
rd.fit(x_train , y_train)
```

```python
rd.best_params_
```

```python
gd = GridSearchCV(DecisionTreeRegressor(),param_grid=df)
gd.fit(x_train,y_train)
```

```python
gd.best_params_
```

```python
gd.best_score_
```
