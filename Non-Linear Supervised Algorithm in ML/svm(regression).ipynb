```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
# CSV file read karna
dataset = pd.read_csv(r"D:\py_machine_learning\datasets\polynomial.csv")

# Data ke first 5 rows dekhna
dataset.head()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Level", y="Salary", data=dataset)
plt.show()
```

```python
x = dataset[["Level"]]
y = dataset["Salary"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import SVR
sv = SVR(kernel='linear',degree=3)
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sv.score(x_train,y_train)*100
```

```python

sns.scatterplot(x="Level", y="Salary", data=dataset)
plt.scatter(dataset["Level"],sv.predict(x),color="red")
plt.show()
```
