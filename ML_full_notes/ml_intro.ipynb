What is ML?  
Ans.- Machine Learning is a field of Artificial Intelligence where computers learn from data and make predictions or decisions without being explicitly programmed.  

How ML Works  

1. Collect data  
2. Clean and preprocess the data  
3. Train a model  
4. Make predictions  
5. Improve the model if predictions are wrong  

### Types of Machine Learning  

### 1. Supervised Learning  
    Supervised Learning Machine Learning ka ek type hai jisme  
    model ko aisa data diya jata hai jisme Input aur Output dono  
    already known hote hain.  
Matlab model ko training ke time pe sahi answer pata hota hai, aur wo usse learn karta hai.  

-- How to work?  
1. Model ko labeled data diya jata hai 
2. Model prediction karta hai  
3. Prediction ko correct output se compare kiya jata hai  
4. Model galti dekh kar improve hota hai  
5. Repeat hota rehta hai until model becomes accurate             

### 2. Unsupervised Learning   
    Unsupervised Learning Machine Learning ka ek type hai jisme  
    model ko sirf Input data diya jata hai, aur koi Output/Label   
    nahi diya jata.Matlab model ko sahi jawab nahi pata hota, phir  
    bhi wo data ke pattern aur structure ko khud discover karta hai.

-- How to work?
1. Model ko raw/unlabeled data diya jata hai
2. Model data ke similarities aur patterns dhundta hai
3. Model data ko groups, clusters, ya structure me organize karta hai
4. Human ko bata deta hai ki data ka actual hidden structure kya hai  

### 3. Reinforcement Learning  
    Reinforcement Learning (RL) Machine Learning ka ek type hai jisme  
    model trial-and-error se seekhta hai.  
    Model kisi environment me actions leta hai, aur uske badle me reward  
    ya penalty milti hai. Model ka goal hota hai maximum reward paana.

-- How to work?
1. Model environment me action leta hai  ?
2. Environment reward ya punishment deta hai  
3. Model samajhta hai ki kya sahi tha kya galat  
4. Next time action improve karta hai
5. Repeat hota rehta hai until model becomes exper

### 4. Semi-Supervised Learning  
    Semi-Supervised Learning Machine Learning ka wo type hai jisme training  
    ke liye thoda sa labeled data aur bohot saara unlabeled data use hota hai.

-- How to work?

1. Model unlabeled data me structure/pattern find karta hai  
2. Thoda sa labeled data se correct output samajhta hai
3. Dono information combine karke prediction karta hai
4. Labeled + unlabeled learning se model zyada accurate ban jata hai
