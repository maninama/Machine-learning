```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\py_machine_learning\datasets\covid_toy - covid_toy.csv")
df
```

```python
df.isnull().sum()
```

```python
df['fever'] = df['fever'].fillna(df['fever'].mean())
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns=['has_covid']) # Input and independant data 
x.shape
```

```python
y = df['has_covid'] # Target and dependent data 
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print("Total Data shape :",df.shape)
print("-----------------------------")
print("Input data shape : ",x.shape)
print("x_train data shape : ",x_train.shape)
print("x_test data shape : ", x_test.shape)
print("-----------------------------")
print("Traget data shape : ",y.shape)
print("y_train data shape : ",y_train.shape)
print("y_test data shape : ", y_test.shape)
```
