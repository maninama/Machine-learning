```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(df.drop(columns=['total_bill',]),df['total_bill'],test_size=0.2,random_state=42)
```

```python
oe = OrdinalEncoder(categories=[['Dinner','Lunch']])
x_train_time = oe.fit_transform(x_train[['time']])


x_test_time = oe.fit_transform(x_test[['time']])

x_train_time.shape
```
