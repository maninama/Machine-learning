```python
import numpy as np 
import pandas as pd 
```

```python
df = pd.read_csv(r"D:\py_machine_learning\datasets\covid_toy - covid_toy.csv")
df.head(2)
```

```python
df = df.dropna()
```

```python
df = df.drop(columns=['age','fever'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

```python
# gender ---> 2-1 = 1
# cough ----> 2-1 = 1
# city -----> 4-1 = 3
# has_covid --->2-1 = 1
```

```python
ohe.fit_transform(df[['gender','cough','city','has_covid']])
```

```python
#get_dummies 
```

```python
a = pd.get_dummies(df,columns=['gender','cough','city','has_covid'],drop_first=True)
```

```python
a = a.astype(int) 
```
