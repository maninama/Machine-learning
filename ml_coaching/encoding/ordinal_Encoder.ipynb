2. Ordinal --> it will encode more than 1 columns.  

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"D:\py_machine_learning\datasets\covid_toy - covid_toy.csv")
df.head()
```

```python
df = df.dropna()
```

```python
df = df.drop(columns=['age','fever'])
```

```python
df.head(2)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Male','Female'],
                    ['Mild','Strong'],
                    ['Kolkata','Bangalore','Delhi','Mumbai'],
                    ['Yes','No']])
```

```python
df_arr = oe.fit_transform(df)
```

```python
df_new = pd.DataFrame(df_arr,columns=df.columns)
```
