Encoding ----> transform your categorical data into numerical data.  
1. Label ----> it will encode data 1-1 columns.  
2. Ordinal --> it will encode more than 1 columns.  
3. OneHot ---> it will create new column based on sub categories but it has only values not column name.  
4. get_dummies -> it will create new column based on sub-categories but it show column name with values.  


```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv(r"D:\py_machine_learning\datasets\toyota - toyota.csv")
df.head(2)
```

```python
x = df.drop(columns=['price'])
y = df['price']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
x['model']= lb.fit_transform(x['model'])
x['transmission']= lb.fit_transform(x['transmission'])
x['fuelType']= lb.fit_transform(x['fuelType'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_sc,columns= x_train.columns)
```

```python
np.round(x_train_new.describe(), 2)
```
