```python
import pandas as pd
```

```python
dataset = pd.read_csv(r"D:\py_machine_learning\datasets\placement.csv")
dataset.head()
```

```python
x = dataset.drop("package", axis=1)
y = dataset["package"]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
bg = BaggingRegressor(estimator=LinearRegression(),n_estimators=100)
bg.fit(x_train,y_train)
bg.score(x_train,y_train)*100 , bg.score(x_test,y_test)*100
```

```python
rf = RandomForestRegressor(n_estimators=190)
rf.fit(x_train,y_train)
rf.score(x_train,y_train)*100 , rf.score(x_test,y_test)*100
```
