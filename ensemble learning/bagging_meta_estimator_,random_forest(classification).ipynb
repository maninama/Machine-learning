```python
import pandas as pd 
import seaborn as sns 
from sklearn.datasets import make_moons 
import matplotlib.pyplot as plt 
```

```python
x,y = make_moons(n_samples=1000,noise=0.2)
```

```python
df = {"x1":x[:,0],"x2":x[:,1],"y":y}
dataset = pd.DataFrame(df)
```

```python
sns.scatterplot(x="x1",y="x2",data=dataset,hue="y")
plt.show()
```

```python
X = dataset[["x1", "x2"]]  #  Input features
y = dataset["y"]   
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
sv = SVC()
sv.fit(x_train, y_train)

train_acc = sv.score(x_train, y_train) * 100
test_acc = sv.score(x_test, y_test) * 100

print(f"Train Accuracy: {train_acc:.2f}%, Test Accuracy: {test_acc:.2f}%")
```

```python
bg = BaggingClassifier(estimator=SVC(), n_estimators=30)
bg.fit(x_train, y_train)

train_acc = bg.score(x_train, y_train) * 100
test_acc = bg.score(x_test, y_test) * 100

print(f"Train Accuracy: {train_acc:.2f}%, Test Accuracy: {test_acc:.2f}%")
```

```python
rf = RandomForestClassifier(n_estimators=30)
rf.fit(x_train, y_train)

train_acc = rf.score(x_train, y_train) * 100
test_acc = rf.score(x_test, y_test) * 100
print(f"Train Accuracy: {train_acc:.2f}%, Test Accuracy: {test_acc:.2f}%")
```
