```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv(r"D:\py_machine_learning\datasets\placement.csv")
dataset.head()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["package"]
```

```python
from sklearn.model_selection import train_test_split 
x_train , x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)*100, lr.score(x_test,y_test)*100
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
dt.score(x_train,y_train)*100, dt.score(x_test,y_test)*100
```

```python
sv = SVR()
sv.fit(x_train,y_train)
sv.score(x_train,y_train)*100, sv.score(x_test,y_test)*100
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
li = [("lr",LinearRegression()),("dt",DecisionTreeRegressor()),("sv",SVR()),]
```

```python
vc = VotingRegressor(li,weights={10,20,30})
vc.fit(x_train,y_train)
vc.score(x_train,y_train)*100 , vc.score(x_test,y_test)*100
```

```python
df = {"lr":lr.predict(x_test),"dt":dt.predict(x_test),"svr":sv.predict(x_test),"vc":vc.predict(x_test),}
```
